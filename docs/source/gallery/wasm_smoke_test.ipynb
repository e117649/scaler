# OpenGRIS Scaler — Browser smoke test

Installs the WASM client wheel into this Pyodide kernel and verifies the imports work.

**Run the cells below in order.**

```python
import js, piplite
# Build an origin-absolute URL so the request escapes the JupyterLite
# service worker (which only serves /lite/* paths).
wheel_url = (
    js.location.origin
    + '/_static/wasm/opengris_scaler-2.1.10-cp313-cp313-emscripten_4_0_9_wasm32.whl'
)
print('Installing from:', wheel_url)
await piplite.install(wheel_url)
```

```python
import sys, scaler
print("Python platform :", sys.platform)
print("Scaler version  :", scaler.__version__)
```

```python
# The pure-Python wasm shim should be loaded under sys.platform == 'emscripten'.
from scaler.io.ymq import ConnectorSocket
print(ConnectorSocket)
print("send_message_sync :", callable(getattr(ConnectorSocket, 'send_message_sync', None)))
print("recv_message_sync :", callable(getattr(ConnectorSocket, 'recv_message_sync', None)))
```

```python
# Importing the Client class checks the InProcessAgentBridge factory wiring.
# Constructing a Client requires a reachable scheduler (ws:// from the browser),
# so we only verify the import here.
from scaler.client.client import Client, check_browser_runtime
print("Client class    :", Client)
print("JSPI preflight  :", check_browser_runtime)
```
