# Parallel square roots with Scaler

Submit a batch of `math.sqrt` calls to a running Scaler cluster and collect the results.

**Prerequisites**: a Scaler scheduler (and matching object storage server) reachable from this kernel. See the [quickstart](../tutorials/quickstart.rst) for how to start one.

Edit `SCHEDULER_ADDRESS` (and optionally `OBJECT_STORAGE_ADDRESS`) in the next cell, then run all cells.

```python
# Connection settings -- edit these to point at your running cluster.
SCHEDULER_ADDRESS = "tcp://127.0.0.1:8516"

# Optional: override the object-storage address advertised by the scheduler
# (useful when the scheduler is behind Docker/Kubernetes port mapping).
# Leave as None to use whatever the scheduler reports.
OBJECT_STORAGE_ADDRESS = None
```

```python
import math

from scaler import Client

with Client(address=SCHEDULER_ADDRESS, object_storage_address=OBJECT_STORAGE_ADDRESS) as client:
    futures = [client.submit(math.sqrt, value) for value in range(16)]
    results = [future.result() for future in futures]

for value, result in enumerate(results):
    print(f"sqrt({value:2d}) = {result:.4f}")
```
